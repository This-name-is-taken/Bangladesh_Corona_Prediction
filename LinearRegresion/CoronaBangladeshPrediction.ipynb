```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from matplotlib.pyplot import figure
```

```python
data = pd.read_csv('coronainfo3.csv')
#data = pd.read_csv('coronainfo2.csv')
```

```python
data
```

```python
plt.scatter(data['date'], data['total_cases'])
plt.show()
#plt.scatter(data2['date'], data2['total_cases'])
#plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(data[['no']] , data[['total_cases']])
```

```python
int(model.predict([[67]]))
```

```python
datanew = []
for i in range(365):
    x = model.predict([[67+i]])
    datanew.append(int(x))
```

```python
dnewest = []
for i in range(len(data['no'])):
    dnewest.append(int(model.predict([[i]])))
```
